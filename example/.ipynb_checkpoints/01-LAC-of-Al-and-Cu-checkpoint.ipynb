```python
# This example will show how to obtain LAC (mm^-1) using crip.
```

```python
import matplotlib.pyplot as plt
from crip.physics import Atten
import numpy as np
```

```python
energyRange = np.arange(10, 151) # keV
# construct Atten object using crip
Al = Atten.fromBuiltIn('Al')
Cu = Atten.fromBuiltIn('Cu')
```

```python
plt.figure()
plt.plot(energyRange, np.log(Al.mu[energyRange]), label='Al')
plt.plot(energyRange, np.log(Cu.mu[energyRange]), label='Cu')
plt.title('LAC of Al and Cu')
plt.xlabel('Energy (keV)')
plt.ylabel('Log LAC (Log mm^-1)')
plt.legend()
```
